# Feature Engineering
- one hot encoding 
- Imputing missing values

### List all the columns that are currently non-numeric

```python
df.select_dtypes(['object']).columns
```

#### Check if there are any missing values

```python
dummy_list =['sub_grade','home_ownership','verification_status','purpose',
             'addr_state','initial_list_status', 'application_type']
df[dummy_list].isnull().any()
```

#### Apply one hot encoding

```python
df = pd.get_dummies(df, columns=dummy_list, drop_first=True)
```

```python
df.sample(2)
```

```python
df.shape
```

#### Imputing missing values

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
imputer.fit(df)
```

```python
df = pd.DataFrame(imputer.transform(df), columns=df.columns)
df.info()
```

df.isnull().any() generates a boolean array (True if the column has a missing value, False otherwise). You can use it to index into df.columns:

```python
df.columns[df.isnull().any()]
```

```python
df.sample(2)
```

```python
df['charged_off'].value_counts()
```

```python
df['charged_off'].value_counts(normalize=True, dropna=False)
```

Loan-status class is imbalanced. To solve the problem of unbalance class issue there are many techniques that can be applied. For example:

(1) Assign a class weight (2) Use ensemble algorithms with cross-validation (3) Upsample minority class or downsample the majority class

Upsample the minority class
One of the popular techniques for dealing with highly unbalanced data sets is called resampling. Although the technique has proven to be effective in many cases to solve the unbalanced class issue, however, these techniques also have their weaknesses. For example, over-sampling records from the minority class, which can lead to overfitting while removing random records from the majority class, which can cause loss of information.

```python
df_major = df[df.charged_off == 0]
df_minor = df[df.charged_off == 1]
```

```python
df_minor_upsmapled = resample(df_minor, replace = True, n_samples = 1076751, random_state = 101)
```

```python
df_minor_upsmapled = pd.concat([df_minor_upsmapled, df_major])
```

```python
df_minor_upsmapled.charged_off.value_counts()
```

#### Train Test Split

Set X and y variables to the .values of the features and label.

```python
X = df_minor_upsmapled.drop('charged_off',axis=1).values
y = df_minor_upsmapled['charged_off'].values
#X = df.drop('charged_off',axis=1).values
#y = df['charged_off'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Building the Model

```python
X_train.shape
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
```

```python
model = Sequential()

# input layer
model.add(Dense(116,  activation='relu'))

# hidden layer
model.add(Dense(58, activation='relu'))

# hidden layer
model.add(Dense(29, activation='relu'))
       
# output layer
# As this is binary classification we use sigmoid activation 
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=50,
          batch_size=4096,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
losses = pd.DataFrame(model.history.history)
losses[['loss','val_loss']].plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,predictions))
```
